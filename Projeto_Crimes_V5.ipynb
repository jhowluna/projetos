```python
import pandas as pd
from datetime import date
import re
import geopandas as gpd
```

```python
dados = pd.read_csv('dados/dataframe_crimes.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly
import folium

%matplotlib inline
sns.set_palette("Accent")
sns.set_style("darkgrid")
```

```python
from folium.plugins import MarkerCluster
```

```python
base = folium.Map(
    location=[-23.6, -46.6],
    tiles='OpenStreetMap',
    zoom_start=11
)
mc = MarkerCluster()

for index, incidente in dados.iterrows():
    mc.add_child(folium.Marker([incidente['latitude'], incidente['longitude']], 
              popup="<h4> Titulo:</h4><h5>"+str(incidente.titulo)+
                               "<h4> Cidade:</h4><h5>"+str(incidente.nome_dist)+
                               "</h5> <p><h4>Data e Hora:</h4> "+str(incidente.time)+
                               "</h5> <p><h4>Sexo:</h4> "+str(incidente.sexo),
                               tooltip = 'Click me!',
              icon=folium.Icon(color='red',icon='info-sign',))).add_to(base)
base.save('templates/GroupCrimes.html')
```

```python
from folium.plugins import HeatMap
```

```python
lat = dados.latitude.tolist()
lng = dados.longitude.tolist()
base = folium.Map(
    location=[-23.6, -46.6],
    tiles='OpenStreetMap',
    zoom_start=10)
HeatMap(list(zip(lat, lng))).add_to(base)
base.save('templates/heatmapCrimes.html')
```
